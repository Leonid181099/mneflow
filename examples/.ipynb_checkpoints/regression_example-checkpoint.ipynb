# Regression, adjusting training parameters, and cross-validation

All MNEflow models can be used for classification and regression. The type of task can be specified during building mnfelow dataset.

````python
#Handle imports and suppress verbosity
import numpy as np
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import mneflow
import tensorflow as tf
tf.get_logger().setLevel('ERROR')
tf.autograph.set_verbosity(0)
````

In this example we will use simulated dataset where an event-related response, which varies from trial-to-trial in amplitude and latency

````python
datafile = np.load(".\\datasets\\sim_erps_50nA.npz")

X = datafile["X"]
y = datafile["y"]
````

When specifying import options we set 
```target_type='float'```
to indicate that we are predicting continous variable
Also, we split the dataset in 5 folds to perform cross-validation

````python
#Specify import options
import_opt = dict(savepath='.\\datasets\\tfr\\',  # path where TFR files will be saved
                  out_name='sim_regression_example',  # name of TFRecords files
                  fs=500,
                  input_type='trials',
                  target_type='float',
                  n_folds=5,
                  bp_filter=(1., 45.), # band-pass filter
                  scale=True,  # apply baseline_scaling
                  crop_baseline=False,  # remove baseline interval after scaling
                  scale_interval=(0,60),  # indices in time axis corresponding to baseline interval
                  overwrite=True,
                  test_set='holdout')


#write TFRecord files and metadata file to disk

meta = mneflow.produce_tfrecords((X, y), **import_opt)
````

For this task we also use LF-CNN, but we want to adjust hyperparameters e.g. use larger temporal convolution kernels, etc. 

Also, we want to apply l2-penaly for adjustable weights of temporal convolution and de-mixing layres, but use l1-penalty for final dense layer. This can be done by modifying the parameter dictionary. 

Hyperparameter specifications differ depending on the network type. For full list of parameters see e.g. help for 
```mneflow.moodels.LFCNN```

````python
dataset = mneflow.Dataset(meta, train_batch=50)


lf_params = dict(n_latent=32, #number of latent factors
                  filter_length=128, #convolutional filter length in time samples
                  nonlin = tf.nn.relu,
                  padding = 'SAME',
                  pooling = 32,#pooling factor
                  stride = 24, #stride parameter for pooling layer
                  pool_type='avg',
                  model_path = import_opt['savepath'],
                  dropout = .25,
                  l2_scope =  ["tconv", "dmx"],
                  l2=3e-3,
                  l1_scope = [ "fc"],
                  l1=3e-3)


model = mneflow.models.LFCNN(dataset, lf_params)

model.build(optimizer='adam', learn_rate=3e-4)
````


We also want to evaluate model performance with cross-validation. Since our dataset has number of folds specified, we just need to indicate that we want to run corss-validation by providing 
```mode='cv'``` argument to ```model.train()```

Note that during cross-validation the model shuffles the weights instead of full re-initialization. This can affect the results.


````python
cv_losses, sv_metrics = model.train(n_epochs=25, eval_step=50, early_stopping=5, mode='cv')
````

To simplify experimentation, all traning results, hyperparamter values and other model-specific information is stored in trainging log ```import_opt['savepath']```
In our case check ```...\mneflow\datasets\tfr\lfcnn_log.csv```


````python
model.evaluate(meta['test_paths'])

model.compute_patterns()

#%%
selection_criterion = 'output_corr'
f1 = model.plot_patterns('Vectorview-mag', sorting=selection_criterion)
f2 = model.plot_spectra(sorting=selection_criterion, norm_spectra="welch", log=True,)
f3 = model.plot_waveforms(sorting=selection_criterion)
````
